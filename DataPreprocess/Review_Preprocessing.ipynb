# Review DataSet 확인

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

## 원본 데이터 가져오기

```python
reviewDF = pd.read_csv("../yelp_csv\yelp_academic_dataset_review.csv")
reviewDF.head()
```

### 데이터셋 크기 확인

```python
reviewDF.shape
```

# Philadelphia의 Business 가져오기

```python
paDF = pd.read_csv("../yelp_csv\yelp_academic_dataset_business.csv")
paDF = paDF[["business_id", "city", "state", "categories", "latitude", "longitude"]]
paDF = paDF[paDF.city=="Philadelphia"]
paDF = paDF.dropna(subset="categories")
paDF = paDF[paDF["categories"].str.contains("Restaurants|Food")]
paDF
```

# Rating 데이터셋 생성

## 필요한 열만 추출

```python
ratingDF = reviewDF[["user_id", "business_id", "stars", "date"]]
ratingDF.head()
```

## Philadelphia의 Review 추출

```python
joinDF = pd.merge(ratingDF, paDF, left_on="business_id", right_on="business_id", how="inner")
joinDF
```

## PA Rating 데이터셋 생성

```python
ratingDF = joinDF[["user_id", "business_id", "stars", "date"]]
ratingDF.head()
```

## 저장 및 확인

```python
ratingDF.to_csv("philadelphia_rating_raw.csv", encoding="utf-8", index=False)
```

```python
ratingDF = pd.read_csv("philadelphia_rating_raw.csv")
ratingDF
```

# 정수형 User, Business ID 추가

## 정수형 User ID 생성

```python
users = ratingDF.user_id.unique()
userDF = pd.DataFrame(data=users, columns=["originalUserId"])
userDF["newUserId"] = userDF.index + 1
userDF
```

## 정수형 Business ID 생성

```python
businesses = ratingDF.business_id.unique()
businessDF = pd.DataFrame(data=businesses, columns=["originalBusinessId"])
businessDF["newBusinessId"] = businessDF.index + 1
businessDF
```

## Rating 데이터셋에 정수형 인덱스 추가

```python
ratingDF = ratingDF.merge(userDF, left_on="user_id", right_on="originalUserId")
ratingDF.drop(labels="originalUserId", axis=1, inplace=True)

ratingDF = ratingDF.merge(businessDF, left_on="business_id", right_on="originalBusinessId")
ratingDF.drop(labels="originalBusinessId", axis=1, inplace=True)

ratingDF
```

## 정수형이 추가된 데이터 셋 저장 및 확인

```python
ratingDF.to_csv("philadelphia_rating.csv", encoding="utf-8", index=False)
```

```python
ratingDF = pd.read_csv("philadelphia_rating.csv")
ratingDF
```

# 데이터셋 분리

## Rating 데이터셋 분리

```python
# Train : Validation : Test =  6 : 2 : 2
X_train, X_test = train_test_split(ratingDF, test_size=0.40, shuffle=True)
X_validation, X_test = train_test_split(X_test, test_size=0.50, shuffle=True)
```

```python
print("Size of train set:", X_train.shape[0])
print("Size of validation set:", X_validation.shape[0])
print("Size of test set:", X_test.shape[0])
```
