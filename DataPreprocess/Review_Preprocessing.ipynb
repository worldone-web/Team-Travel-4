# Philadelphia의 Business 가져오기

```python
df_b = pd.read_csv("yelp_dataset\csv\yelp_academic_dataset_business.csv")
df_b = df_b[["business_id", "city", "state", "categories"]]
df_b = df_b[df_b.city=="Philadelphia"]
df_b = df_b.dropna(subset="categories")
df_b = df_b[df_b["categories"].str.contains("Restaurants|Food")]
df_b
```

# Philadelphia의 Review 추출

```python
df_join = pd.merge(df_a, df_b, left_on="business_id", right_on="business_id", how="inner")
df_join
```
