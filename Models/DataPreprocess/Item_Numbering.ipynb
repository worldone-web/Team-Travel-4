```python
import pandas as pd
import numpy as np
```

```python
reviewDF = pd.read_csv("/content/drive/MyDrive/Data/philadelphia_rating.csv")
reviewDF.head()
```



---



## 비즈니스 넘버링

```python
businessID = reviewDF[['business_id', 'newBusinessId']].drop_duplicates()

print('shape: ', businessID.shape)
print('min', businessID['newBusinessId'].min())
print('max', businessID['newBusinessId'].max())

businessID.head()
```

```python
businessDF = pd.read_csv("/content/drive/MyDrive/yelp_academic_dataset_business.csv")
businessDF = businessDF[["business_id", "categories", "latitude", "longitude", "name"]]
businessDF.head()
```

```python
result = pd.merge(businessDF, businessID, on='business_id')
result.head()
```

```python
result.to_csv("/content/drive/MyDrive/Data/philadelphia_business.csv", index=False)
```

## 사용자 넘버링

```python
userID = reviewDF[['user_id', 'newUserId']].drop_duplicates()

print('shape: ', userID.shape)
print('min', userID['newUserId'].min())
print('max', userID['newUserId'].max())

userID.head()
```
